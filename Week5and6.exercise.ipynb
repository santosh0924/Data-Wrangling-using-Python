### Chapter 7 

```python
# Load necessary libraries
import numpy as np
import pandas as pd
```

```python
# Read The Metropolitan Museum of Art Open Access CSV
met_objects = pd.read_csv("MetObjects.csv")
met_objects.head()
```

```python
# Get rows and column details
met_objects.shape
```

```python
# Count number of null values in each column
met_objects.isnull().sum()
```

```python
# Replace missing values like spaces with NaN
met_objects = met_objects.replace(" ", np.NaN)
```

#### No rows where selected since each row has some column that has missing value

```python
# Drop only if all the columns have missing data
met_objects.dropna(how='all')
```

#### None of the rows where dropped as some columns don't have any missing data

```python
# Example to filter missing data in 'Credit line column'
met_objects_filtr = met_objects.dropna(subset=['Credit Line'])
```

```python
met_objects_filtr.isnull().sum()
```

#### Above result confirms that missing values in 'Credit Line' is dropped

```python
# Fill in missing data. Here one of the column which as missing value is filled with 0
met_objects_fill = met_objects.fillna({'AccessionYear': 0})
```

```python
# Remove duplicates
# First check to see if any duplicates in all columns
met_objects_unq = met_objects.drop_duplicates() 
```

```python
met_objects_unq.shape
```

```python
# Remove duplicates in Object Number 
met_objects_dup = met_objects.drop_duplicates(subset = 'Object Number')
```

```python
met_objects_dup.shape
```

### Chapter 10

```python
# Grouping with Functions
candy = pd.read_excel("candyhierarchy2017.xlsx")
candy.head()
```

```python
candy.shape
```

```python
candy_subs1 = candy[['Internal ID', 'Q2: GENDER', 'Q3: AGE', 'Q4: COUNTRY', 'Q5: STATE, PROVINCE, COUNTY, ETC', 
                     'Q12: MEDIA [Science]']]
```

```python
candy_renm = candy_subs1.rename(columns={"Q2: GENDER": "Gender", "Q3: AGE": "Age", "Q4: COUNTRY": "Country", 
                                         "Q5: STATE, PROVINCE, COUNTY, ETC": "St_Pr_Cty", "Q12: MEDIA [Science]":"Media"})
candy_renm
```

```python
candy_filtr = candy_renm.dropna(subset=['Age'])
candy_filtr
```

```python
candy_filtr.Age = pd.to_numeric(candy_filtr['Age'], errors='coerce')
candy_filtr
```

```python
candy_filtr.groupby('Gender').Age.count()
```

```python
# Split/Apply/Combine
candy_filtr['Age'].groupby(candy_filtr['Gender']).describe()
```

```python
# Cross tabs
# Crosstab example with Gender in x-axis and Country in y-axis
candy_cross = pd.crosstab(candy_filtr.Gender, candy_filtr.Country, margins=True)
candy_cross
```
