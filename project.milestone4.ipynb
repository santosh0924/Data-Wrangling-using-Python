### Milestone 4: Connecting to an API/Pulling in the Data and Cleaning/Formatting

```python
# Import necessary libraries
import urllib.request, urllib.parse, urllib.error
from urllib.error import HTTPError, URLError
import requests
import json
import pandas as pd
from fuzzywuzzy import fuzz 
from fuzzywuzzy import process
```

```python
# Connect to the API and first get all makes for Vehicle type - Car
url = 'https://vpic.nhtsa.dot.gov/api/vehicles/GetMakesForVehicleType/car?format=json'
    
try:
    uh = urllib.request.urlopen(url)        
except HTTPError as e:                      
    print("Sorry! Could not retrieve anything from API")
except URLError as e:                       
    print('Failed to reach a server.')
    print('Reason: ', e.reason)
else:
    carapi = uh.read().decode()           
    print("Successfully retrieved data from API")
```

```python
carapi
```

```python
json_data = json.loads(carapi)
```

```python
dict1 = {'Make':[]}

for d in json_data['Results']:
    dict1['Make'].append(d['MakeName'])
```

```python
df = pd.DataFrame.from_dict(dict1)
dfs = df.sort_values(by=['Make'])
```
