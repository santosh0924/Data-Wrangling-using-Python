```python
# Function to get table data
def table_data(df):
    df_a = df['Automobile'].drop_duplicates().str.split(n=1).str[0]
    df_b = df['Automobile'].drop_duplicates().str.split(n=1).str[1]

# Automobile column as Make and Model combined. Below is to split them
# Replace headers from Automobile to Make and Model

    df.insert(loc=0, column='Make', value=df_a)
    df.insert(loc=1, column='Model', value=df_b)
    
    return df
```
