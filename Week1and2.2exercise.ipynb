```python
"""
1. Create a list of 100 random numbers.
"""
limit=100
random_num_list = [random.randint(0, limit) for x in range(0,limit)]
```

```python
"""
2. Create a new list from this random list, with numbers that are divisile by 3
"""
list_divisible_by_3 = [a for a in random_num_list if a%3 == 0]
```

```python
"""
3. Calculate the length of these two lists and store the difference in a new variable.
"""
length_random_list = len(random_num_list)
length_list_divis_3 = len(list_divisible_by_3)
difference = length_random_list - length_list_divis_3
```

```python
"""
4. Using a loop, perform steps 2 and 3 and find the difference variable three times.
"""
num_experiments = 3
difference_list = []

for i in range(0, num_experiments):
    rand_num_list = [random.randint(0, limit) for x in range(0,limit)]
    list_div_by_3 = [a for a in rand_num_list if a%3 == 0]
    len_rand_list = len(rand_num_list)
    len_list_div_3 = len(list_div_by_3)
    difference = len_rand_list - len_list_div_3
    difference_list.append(difference)
difference_list
```

```python
"""
5. Find the arithmetic mean of these three difference values.
"""
arith_mean = sum(difference_list) / float(len(difference_list))
arith_mean
```

### Activity 4: Design Your Own CSV Parser

```python
"""
1. Import zip_longest from itertools. Create a function to zip header, line and fillvalue=None.
"""
```

```python
def return_dict_from_csv_line(header, line):
    # zip them
    zipped_line = zip_longest(header, line, fillvalue=None)
    # use dict 
    ret_dict = {kv[0]: kv[1] for kv in zipped_line}
    return ret_dict
```

```python
"""
2. Open the accompanying sales_record.csv file from the GitHub link by using r mode inside a with block and first check that
it is opened.
"""
```

```python
with open("sales_record.csv", "r") as fd:
    first_line = fd.readline()
    header = first_line.replace("\n", "").split(",")
    for i, line in enumerate(fd):
        line = line.replace("\n", "").split(",")
        d = return_dict_from_csv_line(header, line)
        print(d)
        if i > 10:
            break
```
